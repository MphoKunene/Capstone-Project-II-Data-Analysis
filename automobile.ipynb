```python
# Importing the required packages here

import numpy as np
import pandas as pd
import seaborn as sns
import ast, json
import missingno

from datetime import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
automobile_df = pd.read_csv('automobile.txt')
```

```python
automobile_df.head()
automobile_df.replace('?', np.NaN)
```

```python
# Plot graphic of missing values
missingno.matrix(automobile_df, figsize = (30,10))

# Seems like there is no missing data from the dataset
```

```python
# Checking for duplicated data
automobile_df.duplicated().sum()

# Seems there is no duplicated data
```

```python
# Replacing "?" with NaN to indicate that the values there are not numbers
automobile_df.replace('?', np.NaN)
```

```python
# for discretised continuous variables
df_auto = automobile_df
```

```python
# How many makes of the car
fig = plt.figure(figsize=(9,10))
sns.countplot(y='make', data=df_auto);
print(df_auto.make.value_counts())
```

As you can see there is alot of toyota makes than there is any other cars while mercury being the least made car

```python
#  creating a temporary dataset
df = automobile_df

df.symboling.plot(kind='hist')
```

As we can see here that 0 symbol is the highest meaning that the cars with the risk factor zero 0 are pretty safe but between 2 to 3 simply means that car is very risky, however between -2 and -1 shows that those car risk factor is pretty low compared to the others.

```python
df.groupby('aspiration').symboling.value_counts(normalize= True).unstack().plot(kind='bar')
```

So I decided to compare the standard and turbo aspirations to see the which symboling risk factor do they fall under 3 being super risky and -2 being less risky.
